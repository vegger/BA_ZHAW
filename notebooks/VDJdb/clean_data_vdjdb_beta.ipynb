```python
import pandas as pd
import numpy as np
import tidytcells as tt
```

```python
from_path = "../../data/VDJdb/fitted_data/vdjdb_fitted_data_beta.tsv"
vdjdb_fitted_beta_df = pd.read_csv(from_path, sep="\t")
vdjdb_fitted_beta_df
```

```python

for new_column in ['TRAV', 'TRAJ', 'TRA_CDR3','TRA_leader', 'TRB_leader', 'TRAC', 'TRBC', 'Linker', 
                   'Link_order', 'TRA_5_prime_seq', 'TRA_3_prime_seq', 'TRB_5_prime_seq', 'TRB_3_prime_seq']:
    vdjdb_fitted_beta_df[new_column] = np.nan 

template_columns_order = [
    'TCR_name', 'TRAV', 'TRAJ', 'TRA_CDR3', 'TRBV', 'TRBJ', 'TRB_CDR3',
    'TRAC', 'TRBC', 'TRA_leader', 'TRB_leader', 'Linker', 'Link_order',
    'TRA_5_prime_seq', 'TRA_3_prime_seq', 'TRB_5_prime_seq', 'TRB_3_prime_seq'
]
vdjdb_beta_schema_df = vdjdb_fitted_beta_df[template_columns_order]
vdjdb_beta_schema_df
```

```python
vdjdb_fitted_beta_df.columns
```

```python
vdjdb_beta_cleaned_df = pd.concat([vdjdb_beta_schema_df, vdjdb_fitted_beta_df["Epitope"]], axis=1)
vdjdb_beta_cleaned_df.columns
```

```python
vdjdb_fitted_beta_df.columns
```

```python
vdjdb_beta_cleaned_df["Score"] = vdjdb_fitted_beta_df["Score"]
vdjdb_beta_cleaned_df["MHC A"] = vdjdb_fitted_beta_df["MHC A"]
vdjdb_beta_cleaned_df["MHC B"] = vdjdb_fitted_beta_df["MHC B"]
vdjdb_beta_cleaned_df["MHC class"] = vdjdb_fitted_beta_df["MHC class"]
vdjdb_beta_cleaned_df
```

```python
def standardize_gene(x):
    if pd.notna(x): 
        return tt.tr.standardize(gene=x, species="homosapiens")
    else:
        return x 

columns_gene = ["TRAV", "TRBV", "TRAJ", "TRBJ"]

for col in columns_gene:
    vdjdb_beta_cleaned_df[col] = vdjdb_beta_cleaned_df[col].apply(standardize_gene)

# vdjdb_paired_cleaned_df[["TRAV", "TRBV", "TRAJ", "TRBJ"]]
```

```python
def standardize_CDR3(x):
    if pd.notna(x): 
        return tt.junction.standardize(seq=x)
    else:
        return x 
    
columns_CDR3 = ["TRA_CDR3", "TRB_CDR3"]

for col in columns_CDR3: 
    vdjdb_beta_cleaned_df[col] = vdjdb_beta_cleaned_df[col].apply(standardize_CDR3)


# vdjdb_paired_cleaned_df[["TRA_CDR3", "TRB_CDR3"]]
```

```python
def standardize_MHC(x): 
    if pd.notna(x): 
        return tt.mh.standardize(gene=x, species="homosapiens")
    
columns_MHC = ["MHC A", "MHC B"]

for col in columns_MHC: 
    vdjdb_beta_cleaned_df[col] = vdjdb_beta_cleaned_df[col].apply(standardize_MHC)

# print(vdjdb_paired_cleaned_df[["MHC A", "MHC B"]].to_string())
```

```python
to_path = "../../data/VDJdb/cleaned_data"
file_name = "vdjdb_cleaned_data_beta.tsv"
vdjdb_beta_cleaned_df.to_csv(to_path+"/"+file_name, sep="\t", index=False)
```
