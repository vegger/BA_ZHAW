```python
def need_to_know(name, df):
  all_count = len(df)
  negative_count = len(df[df["Binding"] == 0])
  positive_count = len(df[df["Binding"] == 1])
  negative_ratio = 100/all_count*negative_count
  positive_ratio = 100/all_count*positive_count
  print(name)
  print(f"positive count:  {positive_count}")
  print(f"positive ratio:  {positive_ratio}&")
  print(f"negative count:  {negative_count}")
  print(f"negative ratio:  {negative_ratio}%")
  print(f"rows all:        {all_count}")
  print()

need_to_know("df_paired_gene_test", df_paired_gene_test)
need_to_know("df_paired_gene_validation", df_paired_gene_validation)
need_to_know("df_paired_gene_train", df_paired_gene_train)

need_to_know("df_paired_allele_test", df_paired_allele_test)
need_to_know("df_paired_allele_validation", df_paired_allele_validation)
need_to_know("df_paired_allele_train", df_paired_allele_train)

need_to_know("df_beta_gene_test", df_beta_gene_test)
need_to_know("df_beta_gene_validation", df_beta_gene_validation)
need_to_know("df_beta_gene_train", df_beta_gene_train)

need_to_know("df_beta_allele_test", df_beta_allele_test)
need_to_know("df_beta_allele_validation", df_beta_allele_validation)
need_to_know("df_beta_allele_train", df_beta_allele_train)
  
```
