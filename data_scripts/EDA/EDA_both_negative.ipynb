```python
def mf_entries(dataset_name, df, columns_of_interest):
  # Group by the columns to check and filter groups with more than one unique 'Binding' value
  grouped = df.groupby(columns_of_interest)
  count_of_entries = 0
  affected_rows = 0
  affected_negative_rows = 0

  for name, group in grouped:
    if len(group[group["Binding"] == 0]) > 0 and len(group[group["Binding"] == 1]) > 0:
      # print("GEFUNDEN")
      # print(name)
      # print()
      count_of_entries += 1
      affected_rows += len(group)
      affected_negative_rows += len(group[group["Binding"] == 0])

  print(f"{dataset_name}:")
  print(f"count of entries:  {count_of_entries}")
  print(f"affected rows:     {affected_rows}")
  print(f"affected neg rows: {affected_negative_rows}")
  print()

mf_entries("paired gene", df_paired_gene_duplicates, columns_of_interest_paired)
mf_entries("paired allele", df_paired_allele_duplicates, columns_of_interest_paired)
mf_entries("beta gene", df_beta_gene_duplicates, columns_of_interest_beta)
mf_entries("beta allele", df_beta_allele_duplicates, columns_of_interest_beta)

# grouped[grouped["TRB_CDR3"] == 'CASINYGAGLNTEAFF']
```
